```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path='/content/drive/My Drive/googleplaystore.csv'
ds1= pd.read_csv(file_path)
ds1
```

```python
print(ds1.isnull().sum())
```

```python
# This removes missing values across the rows (axis=0)
new_data = ds1.dropna(axis=0, how='any')
new_data
```

```python
new_data.dtypes
```

```python
# Extract numeric part from Size column which contains Mb values alone & multiply it with 1000

def convert_size(size_str):
    if isinstance(size_str, str):  # Check if it's a string
        if 'M' in size_str:
            return float(size_str.replace('M', '')) * 1000
        elif 'k' in size_str:
            return float(size_str.replace('k', ''))
    return size_str  # Return as is if not a string or doesn't match

# Apply the conversion function to the 'Size' column
new_data['new_Size'] = new_data['Size'].apply(convert_size)
print(new_data)
```

```python
# To remove $ from price field & convert it to numeric
new_data['Price_Numeric'] = new_data['Price'].astype(str).str.replace('$', '').astype(float)
```

```python
# To convert string type of Reviews to integer
new_data['reviews_new']=new_data['Reviews'].astype(int)
new_data.head(25)
```

```python
# To cut-off + (10,000+) & comma & converting it to type int
new_data['Installs_new']=new_data['Installs'].astype(str).str.replace('+','').str.replace(',','').astype(int)
```

```python
# To convert string type of Reviews to integer and Installs to integer first for comparison
new_data['Reviews_int'] = pd.to_numeric(new_data['Reviews'].astype(int))
new_data['Installs_new_int'] = new_data['Installs_new'].astype(float)

if (new_data['Reviews_int'] > new_data['Installs_new_int']).any():
  new_data_disp = new_data.drop('Reviews', axis=1)
  print(new_data_disp)
else:
  print("No rows found where Reviews are greater than Installs.")
```

```python
dropped_installs_data = new_data[new_data['Installs_new_int'].isnull()]
print("Rows where Installs could not be converted to a number:")
display(dropped_installs_data)
```

```python
new_data['Price_Numeric'] = new_data['Price'].astype(str).str.replace('$', '').astype(float)
```

```python
new_data_disp.dtypes
```

```python
rows_to_drop_indices = new_data[(new_data['Type'] =='Free') & (new_data['Price_Numeric'] > 0)].index
new_data_cleaned = new_data.drop(rows_to_drop_indices)
print("\nCleaned DataFrame (first 7 rows, after dropping):")
print(new_data_cleaned.head(7).to_markdown(index=False, numalign="left", stralign="left"))

print(f"\nNumber of rows before dropping: {len(new_data)}")
print(f"Number of rows after dropping: {len(new_data_cleaned)}")
```

```python
#    This is crucial for accurate string comparison.
new_data['Type'] = new_data['Type'].astype(str).str.strip()
print("\nUnique values in 'Type' column after stripping spaces:")
print(new_data['Type'].unique())
print("\nValue counts for 'Type' column after stripping spaces:")
print(new_data['Type'].value_counts())

condition = (new_data['Type'] == 'Free') & (new_data['Price_Numeric'] > 0)

print("\nValue counts for the combined dropping condition:")
print(condition.value_counts())

print("\nRows identified by the condition (these are the ones that should be dropped):")
print(new_data[condition].to_markdown(index=False, numalign="left", stralign="left"))

rows_to_drop_indices = new_data[condition].index
new_data_cleaned = new_data.drop(rows_to_drop_indices)

print("\n--- Final Results ---")
print("Cleaned DataFrame (first 7 rows, after dropping):")
print(new_data_cleaned.head(7).to_markdown(index=False, numalign="left", stralign="left"))

print(f"\nNumber of rows before dropping: {len(new_data)}")
print(f"Number of rows after dropping: {len(new_data_cleaned)}")
```
